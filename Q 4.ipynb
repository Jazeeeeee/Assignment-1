```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
image = cv.imread('car.jpg')
img = image
STRIDE = (3,5,7)
rows,cols,c = img.shape

f, axarr = plt.subplots(1, len(STRIDE))
for i,S in enumerate(STRIDE):
    img_mod = img.copy()
    for r in range(int(S/2),rows,S):
        for c in range(int(S/2),cols,S):
            img_mod[r-int(S/2):r+int(S/2)+1,c-int(S/2):c+int(S/2)+1] = np.average(img[r-int(S/2):r+int(S/2)+1,c-int(S/2):c+int(S/2)+1])
    axarr[i].imshow(img_mod);
    axarr[i].set_title('{}x{} average block'.format(S,S))
```
