```python
import numpy as np
import cv2 as cv
import math 
from matplotlib import pyplot as plt
```

```python
image = cv.imread('car.jpg')
blurImage1 = cv.blur(image,(3,3)) 
blurImage2 = cv.blur(image,(10,10)) 
blurImage3 = cv.blur(image,(20,20))

kernel1=np.array([[-1,-1,-1],
                 [-1,8,-1],
                 [-1,-1,-1]])
Image20=cv.filter2D(src=image, ddepth=-1, kernel=kernel1)
plt.subplot(321),plt.imshow(image),plt.title('Original')
plt.subplot(322),plt.imshow(blurImage1),plt.title('3x3')
plt.subplot(323),plt.imshow(blurImage2),plt.title('10x10')
plt.subplot(324),plt.imshow(blurImage3),plt.title('20x20')
plt.subplot(325),plt.imshow(Image20),plt.title('Sharpened Image')
```
